```python
# python visualization libraries
import pandas as pd
import numpy as np

import math
import json


import plotly.express as px
import plotly.graph_objects as go

# for hierarchical data
import networkx as nx
```

```python
# ref: https://stackoverflow.com/questions/32400867/pandas-read-csv-from-url

url = 'https://raw.githubusercontent.com/a95123hsu/IST526-Assignment5/main/IST%20526%20TREE%20MAP%20-%20final.csv'

# pandas call to read csv file 
df = pd.read_csv(url)

# quickly show the dataframe
df.head()
# it's a good idea to peek at the tail too. 
# Note, we need display(.) function if more than output is printed
display(df.head())
display(df.tail())

# get column names
display(df.columns)
# remove NaN with blank, otherwise plotly will be upset
df.fillna('', inplace = True)
# df.dropna(axis=0, inplace = True)
display(df.head())
```

```python
fig = px.treemap(
  df,   
  parents = 'Parent',
  names = 'Research Area and Faculty Name',
  values='Number of Publications'
  #color='Indent Level'
)

fig.update_layout(
  title_text="Treemap Diagram", 
  font_size=12
)

fig.show()
```
